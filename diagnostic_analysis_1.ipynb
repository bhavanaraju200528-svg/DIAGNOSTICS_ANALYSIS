```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt
member_names=["aman","riya","karan","sneha","rahul"]
ages = [22,25,26,27]
workout_hours = [2,1,5,2.5,1,3]
colories_burned = [500,600,700,800,900]
print(" finesss center dataset")
l = len(member_names)
for i in range(l-1):
    print(member_names[i],ages[i],workout_hours[i],colories_burned[i])
#mean workout hours
average_workout = sum(workout_hours)/len(workout_hours)
print("\n average workout hours =", average_workout)
#max calories
max_calories = max(colories_burned)
print("\n max calories =", max_calories)
#min calories
min_calories = min(colories_burned)
print("\n min calories =", min_calories)
# median
median_calories = statistics.median(colories_burned)
print("\n median calories =", median_calories)
#mode
mode_calories = statistics.mode(colories_burned)
print("\n mode calories =", mode_calories)

std_dev = statistics.stdev(colories_burned)
print("\n standard deviation =", std_dev)

index = colories_burned.index(max_calories)
print("most active member",member_names[index])
index = colories_burned.index(min_calories)
print("least active member",member_names[index])

plt.bar(member_names,colories_burned)
plt.xlabel("member names")
plt.ylabel("colories burned")
plt.title("fitness center")
plt.show()
```

```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt

# Read uploaded CSV file
df = pd.read_csv('/content/gym_members_exercise_tracking_transformed.csv')

# Display first 5 rows
print("Fitness Center Dataset")
print(df.head())

# Extract required columns
member_names = df['Gender']          # You can replace with Name column if available
ages = df['Age']
df['Session_Duration (hours)'] = df['Session_Duration (hours)'].astype(int)
calories_burned = df['Calories_Burned']

# Mean workout hours
average_workout = workout_hours.mean()
print("\nAverage workout hours =", average_workout)

# Max calories
max_calories = calories_burned.max()
print("\nMax calories =", max_calories)

# Min calories
min_calories = calories_burned.min()
print("\nMin calories =", min_calories)

# Median calories
median_calories = statistics.median(calories_burned)
print("\nMedian calories =", median_calories)

# Mode calories
mode_calories = statistics.mode(calories_burned)
print("\nMode calories =", mode_calories)

# Standard deviation
std_dev = statistics.stdev(calories_burned)
print("\nStandard deviation =", std_dev)

# Most active member
index = calories_burned.idxmax()
print("\nMost active member:")
print(df.loc[index])

# Least active member
index = calories_burned.idxmin()
print("\nLeast active member:")
print(df.loc[index])

# Bar chart
plt.figure(figsize=(10,5))
plt.bar(member_names[:10], calories_burned[:10])

plt.xlabel("Gender")
plt.ylabel("Calories Burned")
plt.title("Fitness Center Analysis of genders and calories burned")
plt.show()
# Bar chart
plt.figure(figsize=(10,5))
plt.bar(member_names[:10], workout_hours[:10])

plt.xlabel("Gender")
plt.ylabel("workout hours ")
plt.title("Fitness Center Analysis of gender and workout hours")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# READ DATASET FROM CSV FILE LINK


df = pd.read_csv('/content/Sales Transaction v.4a.csv')

print("SALES TRANSACTION DATASET\n")

# selecting columns
product = df['ProductName']
price = df['Price']
quantity = df['Quantity']
country = df['Country']

# length
l = len(df)



print("\nDIAGNOSTIC ANALYSIS\n")

sat = 0
nsat = 0

# diagnostic analysis
for i in range(10):

    print(product[i], price[i], quantity[i], country[i])

    if (price[i] > 10 and quantity[i] > 5):

        print("Sales performance is GOOD\n")
        sat = sat + 1

    else:

        print("Sales performance is BAD\n")
        nsat = nsat + 1

# BAR GRAPH
plt.bar(["Good Sales", "Bad Sales"], [sat, nsat])

plt.xlabel("Sales Status")
plt.ylabel("Count")
plt.title("Sales Diagnostic Analysis")

plt.show()

# COUNTRY WISE ANALYSIS
print("\nCOUNTRY ANALYSIS\n")

country_count = df['Country'].value_counts()

print(country_count)

# COUNTRY GRAPH
country_count.head(5).plot(kind='bar')

plt.xlabel("Country")
plt.ylabel("Transactions")
plt.title("Top Countries by Sales")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# READ DATASET
df = pd.read_csv('/content/ecommerce_dataset_updated.csv')

print("ECOMMERCE DATASET\n")

# SELECTING COLUMNS
category = df['Category']
price = df['Price (Rs.)']
discount = df['Discount (%)']
final_price = df['Final_Price(Rs.)']
payment_method = df['Payment_Method']



print("\nDIAGNOSTIC ANALYSIS\n")

sat = 0
nsat = 0

# DIAGNOSTIC ANALYSIS
for i in range(10):

    print(category[i], price[i], discount[i], final_price[i], payment_method[i])

    # CONDITIONS
    if (price[i] > 200 and discount[i] >= 20):

        print("Customer purchase is GOOD\n")
        sat = sat + 1

    else:

        print("Customer purchase is LOW\n")
        nsat = nsat + 1

# BAR GRAPH
plt.bar(["Good Purchase", "Low Purchase"], [sat, nsat])

plt.xlabel("Purchase Status")
plt.ylabel("Count")
plt.title("Ecommerce Diagnostic Analysis")

plt.show()

# CATEGORY ANALYSIS
print("\nCATEGORY ANALYSIS\n")

category_count = df['Category'].value_counts()

print(category_count)

# CATEGORY GRAPH
category_count.plot(kind='bar')

plt.xlabel("Category")
plt.ylabel("Number of Purchases")
plt.title("Category Wise Purchases")

plt.show()

# PAYMENT METHOD ANALYSIS
print("\nPAYMENT METHOD ANALYSIS\n")

payment_count = df['Payment_Method'].value_counts()

print(payment_count)

# PIE CHART
payment_count.plot(kind='pie', autopct='%1.1f%%')

plt.title("Payment Method Distribution")

plt.ylabel("")

plt.show()
```
